```python
# 13. Implement Fuzzy set operations – union, intersection and complement.
# Demonstrate De Morgan’s Law ( Complement of Union) with 2 fuzzy sets. 
# 14. Demonstrate De Morgan’s Law ( Complement of Intersection) with 2 fuzzy sets. 
# 2 FUZZY SETS
def fuzzy_union(A, B):
    return {x: max(A.get(x, 0), B.get(x, 0)) for x in A.keys() | B.keys()}

def fuzzy_intersection(A, B):
    return {x: min(A.get(x, 0), B.get(x, 0)) for x in A.keys() | B.keys()}

def fuzzy_complement(A):
    return {x: round(1 - A[x], 2) for x in A}

def de_morgan_laws(A, B):
    lhs1 = fuzzy_complement(fuzzy_union(A, B))  # ~(A ∪ B)
    rhs1 = fuzzy_intersection(fuzzy_complement(A), fuzzy_complement(B))  # ~A ∩ ~B

    lhs2 = fuzzy_complement(fuzzy_intersection(A, B))  # ~(A ∩ B)
    rhs2 = fuzzy_union(fuzzy_complement(A), fuzzy_complement(B))  # ~A ∪ ~B

    return lhs1, rhs1, lhs1 == rhs1, lhs2, rhs2, lhs2 == rhs2

A = {}
B = {}

n = int(input("Enter number of elements in Fuzzy Set A: "))
for _ in range(n):
    element, membership = input("Enter element and its membership value for A: ").split()
    A[element] = float(membership)

m = int(input("Enter number of elements in Fuzzy Set B: "))
for _ in range(m):
    element, membership = input("Enter element and its membership value for B: ").split()
    B[element] = float(membership)

# Maintain order of elements from A and B
all_elements = list(A.keys()) + [x for x in B.keys() if x not in A]

print("\nFuzzy Set A:", A)
print("Fuzzy Set B:", B)

union_result = fuzzy_union(A, B)
intersection_result = fuzzy_intersection(A, B)
complement_A = fuzzy_complement(A)
complement_B = fuzzy_complement(B)

print("\nFuzzy Union (A ∪ B):", {x: union_result[x] for x in all_elements})
print("Fuzzy Intersection (A ∩ B):", {x: intersection_result[x] for x in all_elements})
print("Fuzzy Complement (~A):", {x: complement_A[x] for x in A})
print("Fuzzy Complement (~B):", {x: complement_B[x] for x in B})

lhs1, rhs1, is_equal1, lhs2, rhs2, is_equal2 = de_morgan_laws(A, B)
print("\nDe Morgan’s Law Verification:")
print("First Law: ~(A ∪ B) = (~A ∩ ~B)")
print("LHS (Complement of Union):", {x: lhs1[x] for x in all_elements})
print("RHS (Intersection of Complements):", {x: rhs1[x] for x in all_elements})
print("De Morgan’s First Law Holds:", is_equal1)

print("\nSecond Law: ~(A ∩ B) = (~A ∪ ~B)")
print("LHS (Complement of Intersection):", {x: lhs2[x] for x in all_elements})
print("RHS (Union of Complements):", {x: rhs2[x] for x in all_elements})
print("De Morgan’s Second Law Holds:", is_equal2)
# 3 x y z 3 x y z
```

```python
# 12. Implement Fuzzy set operations – union, intersection and complement.
# Demonstrate these operations with 3 fuzzy sets.
#3 FUZZY SETS

def fuzzy_union(A, B, C):
    return {x: max(A.get(x, 0), B.get(x, 0), C.get(x, 0)) for x in A.keys() | B.keys() | C.keys()}

def fuzzy_intersection(A, B, C):
    return {x: min(A.get(x, 0), B.get(x, 0), C.get(x, 0)) for x in A.keys() | B.keys() | C.keys()}

def fuzzy_complement(A):
    return {x: round(1 - A[x], 2) for x in A}

# Input for 3 fuzzy sets
A = {}
B = {}
C = {}

n = int(input("Enter number of elements in Fuzzy Set A: "))
for _ in range(n):
    element, membership = input("Enter element and its membership value for A: ").split()
    A[element] = float(membership)

m = int(input("Enter number of elements in Fuzzy Set B: "))
for _ in range(m):
    element, membership = input("Enter element and its membership value for B: ").split()
    B[element] = float(membership)

p = int(input("Enter number of elements in Fuzzy Set C: "))
for _ in range(p):
    element, membership = input("Enter element and its membership value for C: ").split()
    C[element] = float(membership)

# Collect all elements in order
all_elements = list(A.keys()) + [x for x in B.keys() if x not in A] + [x for x in C.keys() if x not in A and x not in B]

print("\nFuzzy Set A:", A)
print("Fuzzy Set B:", B)
print("Fuzzy Set C:", C)

union_result = fuzzy_union(A, B, C)
intersection_result = fuzzy_intersection(A, B, C)
complement_A = fuzzy_complement(A)
complement_B = fuzzy_complement(B)
complement_C = fuzzy_complement(C)

print("\nFuzzy Union (A ∪ B ∪ C):", {x: union_result[x] for x in all_elements})
print("Fuzzy Intersection (A ∩ B ∩ C):", {x: intersection_result[x] for x in all_elements})
print("Fuzzy Complement (~A):", {x: complement_A[x] for x in A})
print("Fuzzy Complement (~B):", {x: complement_B[x] for x in B})
print("Fuzzy Complement (~C):", {x: complement_C[x] for x in C})
# 2 x1 x2 2 x2 x3 2 x1 x3
```
